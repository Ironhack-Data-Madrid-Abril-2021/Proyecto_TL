**Hypothesis: Shark attacks are common during winter mornings**

I will study the correlation between time frames (hours and months) and shark attacks.

```python
#Import libraries and cleaned csv file
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('attacks_clean.csv')
```

Creation of a bar chart which illustrates the split between fatal and non-fatal attacks. 
We can see that most of the attacks were non-fatal, only a 17% where fatal meaning the victim actually died.

```python
df[df['Fatal (Y/N)'].isin(['Y','N'])]["Fatal (Y/N)"].value_counts()\
.plot(kind='bar',figsize=[8,6], title= "Fatal vs non-fatal attacks", color= ['green','red'])
```

Analyse the monthly distribution of the attacks.
Where we can see that most of the attacks happened in September, August and July. The month that had lowest attacks was November. Thus the hypothesis: Shark attacks are common during winter is null, as they are common in summer time.

```python
df['month'].value_counts(dropna=False)
%matplotlib inline
plt.figure(figsize=[12,6])
graph_month = sns.countplot(x='month', data=df, palette='Blues')
plt.xlabel('Months')
plt.ylabel ('Attacks')
plt.title('Shark attacks - Monthly Basis')
```

Analyse when shark attacks are most likely to happen. 
We see that sharks attacks are most likely to happen during the afternoon, thus the hypothesis Shark attacks are common during winter mornings is null.

```python
df['sessions'].value_counts()
df.drop(df[df['sessions'] == '0'].index, inplace = True)
df['sessions'].value_counts(dropna=False)
%matplotlib inline
plt.figure(figsize=[12,8])
graph_sessions = sns.countplot(x='sessions', data=df, palette='Dark2')
plt.xlabel('Sessions')
plt.ylabel ('Attacks')
plt.title('Shark attacks - Time of the Day')
```

**Conclusion:**
Hypothesis Shark attacks are common during winter mornings is null. 
After analyzing the data we can see that most of shark attacks are non-fatal, and happen in summer afternoons.
